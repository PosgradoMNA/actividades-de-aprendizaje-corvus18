```python
import pandas as pd
import numpy as np

url = 'https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv'

df = pd.read_csv(url)
```

```python
df.head()
```

```python
df.describe()
```

## **Sustituyendo valores faltantes**

```python
countna = df.isna().sum()

countna
```

```python
cond = df['X9'].isnull()

df[cond]
```

```python
df_mean = df['X9'].mean()
df['X9'] = df['X9'].replace(np.nan,df_mean)
```

```python
cond = df['X9'].isnull()

df[cond]
```

## **Formato**

```python
df['X5'].head()
```

```python
df['X5'] = df['X5']*5
```

```python
df['X5'].head()
```

```python
df.rename(columns={'X5':'Nuevo X5'}, inplace=True)
```

```python
df.head()
```

```python
df.dtypes
```

## **Normalizacion**

```python
df["Nuevo X5"]
```

```python
df["Nuevo X5"] = (df["Nuevo X5"] - df["Nuevo X5"].min())/(df["Nuevo X5"].max() - df["Nuevo X5"].min())

df["Nuevo X5"]
```

## **Binning**

```python
ld = np.linspace(df["Nuevo X5"].min(),df["Nuevo X5"].max(),4)
group_names = ["Joven", "Medio", "Viejo"]
nuev = pd.cut(df["Nuevo X5"],ld, labels=group_names,include_lowest = True)
```
