```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

url = 'https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv'

df = pd.read_csv(url)
```

```python
df =df.dropna()
df.describe()
```

```python
df["X5"].value_counts()
```

```python
sns.boxplot(x = "X6", y = "X7", data = df)
```

```python
x = df["X5"]
y = df["X15"]

plt.scatter(x,y)
plt.title("Grafico de dispersin")
plt.xlabel("Edad")
plt.ylabel("Monto Deuda")

```

```python
df_test = df[["X5","X15"]]

gp = df_test.groupby("X5", as_index = False).mean()

gp
```

```python
g_pivot = gp.pivot(index="X5", columns = "X15")
```

## **ANOVA & Correlacin**

```python
#Anova

anova = stats.f_oneway(df['X1'], df['X15'], df['X16'], df['X20'])

anova
```

```python
sns.regplot(x="X5", y ="X21", data = df)
plt.ylim(0,)
```

```python

corr, pval = stats.pearsonr(df["X5"],df["X15"])

print("Correlacion: ", corr)

print("P-Value: ",pval)
```
