```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

url = 'https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv'

df = pd.read_csv(url)
df = df.dropna()
```

```python
df.head()
```

```python
#Regresion Lineal: Relacion entre 2 variables(indepenciente(x) y dependiente(y))
# y = ax + b
```

```python
model = LinearRegression()
reg = model.fit(df[['X12']],df[['X13']])
```

```python
print("Intereccion: ",reg.intercept_,", Coeficiente: ",reg.coef_)
```

```python
sns.regplot(x="X12", y="X13", data=df)
plt.ylim(0,)
```

```python
#Score R cuadrada
reg.score(df[['X12']],df[['X13']])
```

image.png

```python
Input=[('scale',StandardScaler()),('model',LinearRegression())]

pipe=Pipeline(Input)

pipe.fit(df[['X12']],df[['X13']])

ypipe=pipe.predict(df[['X12']])

ypipe
```
